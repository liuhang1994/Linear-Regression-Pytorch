# **Linear Regression With Pytorch**

### In this section we are going to implement a simple linear regression with pytorch 

### Lets import necessary libraries. We try to use pure pytorch without adding numpy library.

```python
from __future__ import print_function
import torch 
import matplotlib.pyplot as plt
%matplotlib inline
```

## First we generate some data and show them with matplotlib.

```python
m , c = 3,5
x = torch.rand(100)
rand_noise = torch.randn(100)*0.2
y = m * x + c +rand_noise 
```

```python
plt.plot(x.numpy(),y.numpy() , 'bo')
plt.ylim(4,10)
plt.show()
```

```python
epoch = 10000
lr = 0.01

w = torch.rand(1)
b = torch.rand(1)
loss = torch.zeros(epoch)
print(loss.size())
```

```python
for i in range(epoch):
    y_hat = w * x + b 
    loss[i] = 0.5 * torch.sum((y - y_hat)**2)
    w = w  + lr * torch.dot(x,(y - y_hat))
    b = b + lr *(torch.sum(y - y_hat))

plt.plot(loss.numpy())    
print(w)     
    
    
```

```python
y_l = w * x + b
plt.plot(x.numpy(),y.numpy(),'bo')
plt.hold
plt.plot(x.numpy() , y_l.numpy() ,'r')
plt.ylim(2,10)
plt.show()
```
