# **Linear Regression With Pytorch**

### In this section we are going to implement a simple linear regression with pytorch 

### Lets import necessary libraries. We try to use pure pytorch without adding numpy library.

```python
from __future__ import print_function
import torch 
import matplotlib.pyplot as plt
%matplotlib inline
```

###  **Generate Data and Show with Matplotlib**

```python
m , c = 3,5
x = torch.rand(100)
rand_noise = torch.randn(100)*0.2  #random noise generated by normal distribution
y = m * x + c + rand_noise          
```

```python
plt.plot(x.numpy(),y.numpy() , 'bo')  # convert x and y to numpy array with numpy() function
plt.xlabel('X')
plt.ylabel('Y')
plt.ylim(4,10)                        # set limit bound to y axis 

plt.show()
```

### **Intialize Hyperparameters**

```python
epoch = 500    # number of steps used to learn w and b 
lr = 0.0005     # learning rate 
```

### Initialize w and b 
\begin{equation}
 \mathbf{W} \times \mathbf{X} + \mathbf{b}
\end{equation}


```python
w = torch.rand(1)  # intialize w with a random value
b = torch.rand(1)  # initialize bias with a random value 
loss = torch.zeros(epoch)  # a torch tensor to save l
print(loss.size())
```

### **Training Loop** 
#### In this section we use gradient decent method to update w and b values in each iteration 

```python
for i in range(epoch):
    y_hat = w * x + b 
    loss[i] = 0.5 * torch.sum((y - y_hat)**2)
    w = w  + lr * torch.dot(x,(y - y_hat))
    b = b + lr *(torch.sum(y - y_hat))

plt.plot(loss.numpy()) 
plt.xlabel('steps')
plt.ylabel('loss')
print(w)    
    
    
```

### ** Draw a Fitted Line using Trained Parameters**

```python
y_l = w * x + b
plt.plot(x.numpy(),y.numpy(),'bo')
plt.hold
plt.plot(x.numpy() , y_l.numpy() ,'r')
plt.ylim(2,10)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
